```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import keras
from tensorflow.keras import layers
from tensorflow.keras import Sequential
from tensorflow.keras import optimizers
from tensorflow.keras.layers import Dense
from tensorflow.keras import regularizers


from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import multilabel_confusion_matrix
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
from sklearn.datasets import make_multilabel_classification
from sklearn.model_selection import KFold
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
#load the data
df = pd.read_csv('sudo model 1.csv')
df.head()
```

```python
df = df[["Result", "PR_Interval", "RR_Interval", "RS_Interval"]]
df.head()
```

```python
#seperate input and output
X,y = df.iloc[:, 1:], df.iloc[:, 0]
#m\change outputs into numbers
le = LabelEncoder()
le.fit(y)
y = le.transform(y)

#seperate training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#define the model
#model = Sequential()
#model.add(Dense(3, activation='relu', use_bias=True, kernel_regularizer = keras.regularizers.l1_l2(l1=0.01, l2=0.01),
#               bias_regularizer='l2', activity_regularizer='l2', input_shape=(3,)))
#model.add(Dense(10, activation='relu', use_bias=True, kernel_regularizer = keras.regularizers.l1_l2(l1=0.01, l2=0.01),
 #              bias_regularizer='l2', activity_regularizer='l2'))
#model.add(Dense(50, activation='relu', use_bias=True, kernel_regularizer = keras.regularizers.l1_l2(l1=0.01, l2=0.01),
#               bias_regularizer='l2', activity_regularizer='l2'))
#model.add(Dense(50, activation='relu', use_bias=True, kernel_regularizer = keras.regularizers.l1_l2(l1=0.01, l2=0.01),
#               bias_regularizer='l2', activity_regularizer='l2'))
#model.add(Dense(10, activation='relu', use_bias=True, kernel_regularizer = keras.regularizers.l1_l2(l1=0.01, l2=0.01),
#               bias_regularizer='l2', activity_regularizer='l2'))
#model.add(Dense(5, activation='softmax'))

model = Sequential()
model.add(Dense(48, activation='relu', use_bias=True, kernel_regularizer = keras.regularizers.l1_l2(l1=0.01, l2=0.01),
               bias_regularizer='l2', activity_regularizer='l2', input_shape=(3,)))

model.add(Dense(24, activation='relu', use_bias=True, kernel_regularizer = keras.regularizers.l1_l2(l1=0.01, l2=0.01),
               bias_regularizer='l2', activity_regularizer='l2'))

model.add(Dense(12, activation='relu', use_bias=True, kernel_regularizer = keras.regularizers.l1_l2(l1=0.01, l2=0.01),
               bias_regularizer='l2', activity_regularizer='l2'))

model.add(Dense(5, activation='softmax'))
```

```python
#complie the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
#fit the model
model.fit(X, y, epochs=100, batch_size=50, verbose=2)
```

```python
# evaluate the model
loss = model.evaluate(X, y, verbose=True)
```

```python
prediction = model.predict(X_test)
print(prediction)
```

```python
y_pred=np.argmax(prediction, axis=1)
cm=confusion_matrix(y_test, y_pred)
print(cm)
```
