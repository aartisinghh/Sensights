```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filenumber=np.array([100, 101, 102, 103, 104, 105, 106, 107, 
                     108, 109, 111, 112, 113, 114, 115, 116, 117, 118, 119, 
                     121, 122, 123, 124, 200, 201, 202, 203, 205, 207, 208, 
                     209, 210, 212, 213, 214, 215, 217, 219, 220, 221, 222, 223, 228, 230, 231, 232, 233, 234])
normalized=np.array([])
for i in filenumber:
    print(i)
    df = pd.read_csv("mitbih_database/"+str(i)+".csv")
    try:
        ecg = df["\'MLII\'"]
    except KeyError:
        ecg = df["\'V5\'"]
    maximum = np.amax(ecg)
    minimum = np.amin(ecg)
    norm=(ecg-minimum)/(maximum-minimum)
    normalized = np.concatenate((normalized, norm))
```

```python
print(len(normalized))
```

```python
plt.figure(figsize=[12, 12])

plt.plot(normalized)
```
